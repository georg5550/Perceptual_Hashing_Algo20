```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
h_type = ['aHash', 'bHash']
i_type = ['8x8', '16x16']
index = [1, 2, 3, 4]

h_time8x8 = [0.458, 0.922]              # Hashing time for 8x8 input, [aHash, pHash]
h_time16x16 = [0.322, 1.1]              # Hashing time for 16x16 input, [aHash, pHash]
ds_time = [0.1, 0.05]                   # Downsampling of input, [8x8, 16x16]
hd_8x8 = [[0, 0], [2, 5], [16, 11]]     # Hamming distance of 8x8 hash outputs, [PicX[aHash, pHash]]
hd_16x16 = [[0, 0], [11, 58], [68, 54]] # Hamming distance of 16x16 hash outputs, [PicX[aHash, pHash]]
```

```python
mean_h = []

for x in range(4):
    mean_h.append((sum(h_time8x8) + sum(h_time16x16)) / 4)   
```

```python
plt.figure(figsize = (5, 5))
plt.bar([index[0], index[2]], h_time8x8, label = '8x8')
plt.bar([index[1], index[3]], h_time16x16, label = '16x16', color = 'green')
plt.plot([0.5, 2, 3, 4.5], mean_h, 'r--')
plt.xticks([])
plt.xlim(0.5, 4.5)
plt.ylim(0, 1.3)

plt.annotate('Mean runtime', (0.6, 0.72))
plt.annotate('aHash', (1.3, 0.5))
plt.annotate('pHash', (3.3, 1.15))

#plt.xlabel('Hashing type', fontsize = 20)
plt.ylabel('Algorithm runtime [s]', fontsize = 20)
plt.title('Algorithm runtime', fontsize = 20)
plt.legend(loc = 'upper left')

plt.savefig('hash_type.png', transparent = True)
plt.show()
```
